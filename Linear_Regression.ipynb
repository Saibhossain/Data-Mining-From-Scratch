# Linear Regression in Machine learning

Linear regression is a type of supervised machine-learning algorithm that learns from the labelled datasets and maps the data points with the most optimized linear functions, which can be used for prediction on new datasets.

* Regression
The term regression is used when you try to find the relationship between variables.
* Independent variable (input): Hours studied because it's the factor we control or observe.
* Dependent variable (output): Exam score because it depends on hobw many hours were studied.


```python
# let's assume a datapoints 

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
# visualized the dataset 

import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.show()
```

##  Equation of the (Simple Linear Regression)

For **simple linear regression**, the model predicts the target using a straight line:

$$
y = mx + b
$$

### ðŸ” Variable Definitions:
- **$ y $**: Predicted value (dependent variable)  
- **$ x $**: Input feature (independent variable)  
- **$ m $**: **Slope** â€” change in $ y $ for a one-unit increase in $ x $  
- **$ b $**: **Intercept** â€” predicted value of $ y $ when $ x = 0 $

> The "best-fit" line is found by choosing $ m $ and $ b $ that **minimize the sum of squared residuals** (ordinary least squares).

```python
# draw a line of Linear Regression
from scipy import stats

m, b, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    y = (m * x) + b
    return  y

mymodel = list(map(myfunc,x))

plt.scatter(x,y)
plt.plot(x, mymodel)
plt.show()
```
